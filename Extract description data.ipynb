### Set the pipeline to split the text in spacy

```python
nlp = spacy.load('pt_core_news_sm')
```

```python
def pipe_spliting(doc):
    start = 0
    for token in doc:
        if token.text == '|':
            yield doc[start:token.i]
            start = token.i
    if start < len(doc):
        yield doc[start:len(doc)]
```

```python
sentence_segmenter = SentenceSegmenter(nlp.vocab, strategy=pipe_spliting)
nlp.add_pipe(sentence_segmenter)
```

```python
####################################################################################
for file, key_words in dic_file_key_word.items(): 
    # print(file, key_word)
    # Reed the json files with the pages data
    with open(f'crawled pages/json/{file}.json', 'r') as f:
        pages_list = json.loads(f.read())['pages']
    # print('pages_list: ', len(pages_list))
    
    # Add the parameter description and set label to 'Not auction object'
    for page in pages_list:
        page['description_from_text'] = ''
        page['label'] = 'Not auction object'
        if 'text_from_tag' not in page:
            page['text_from_tag'] = ''
            
    # Extract the description from the raw text and fill in the 'description' field in the page object
    # p = 0
    for page in pages_list:
        # p += 1
        text = page['text'].lower()
        for key_word in key_words:
            if key_word in text:
                # print(f'Key word found in page: {p}') 
                doc = nlp(text)
                sentences = list(doc.sents)
                for i in range(len(sentences)):
                    if key_word in sentences[i].text:
                        # if the sentence contains only the key word
                        if sentences[i].text.replace('|', '').strip() == key_word:
                            j = 1
                            while sentences[i + j].text.replace('|', '').strip() == '':
                                j += 1
                            sentence = sentences[i + j].text.replace('|', '').strip()
                            page['description_from_text'] = sentence
                            # print(i, sentence)
                        # if the sentence contains the key word and the description
                        else:
                            start = sentences[i].text.find(key_word) + len(key_word) + 1
                            sentence = sentences[i].text[start:].replace('|', '').strip()
                            page['description_from_text'] = sentence
                            # print(f'*{i} {sentence}')
        
                        # Set the text to be analyzed to define the label
                        if page['text_from_tag'] == '':
                            description = page['description_from_text']
                        elif page['text_from_tag'] != '':
                            description = page['text_from_tag']
                        
                        # Search for contex words and set the label
                        label = ''
                        for word in immobile_words:
                            if word in description:
                                label = 'immobile'
                                break
                        if label == '':        
                            for word in car_words:
                                if word in description:
                                    label = 'car'
                                    break
                        if label == '':
                            for word in informatic_words:
                                if word in description:
                                    label = 'informatic'
                                    break
                        if label == '':
                            label = 'other'
                        page['label'] = label

    # Save the file
    pages_with_description_json = json.dumps({'pages_list':pages_list}, ensure_ascii=False).encode(encoding='utf-8')
    with open(f'crawled pages/descriptions/{file}_descriptions.json', 'w') as f:
        f.write(pages_with_description_json.decode())
```
