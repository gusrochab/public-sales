```python
import json
from os import listdir
from os.path import isfile, join
files_descs = [f for f in listdir('/Users/gustavo/Desktop/Gustavo/Projetos TI/public-sales/crawled pages/descriptions2') if isfile(join('/Users/gustavo/Desktop/Gustavo/Projetos TI/public-sales/crawled pages/descriptions2', f))]
files = [f for f in listdir('/Users/gustavo/Desktop/Gustavo/Projetos TI/public-sales/crawled pages/json') if isfile(join('/Users/gustavo/Desktop/Gustavo/Projetos TI/public-sales/crawled pages/json', f))]
```

```python
for file in files:
    print(file)
```

```python
for file_desc in files_descs:
    # Get text descroption
    print(file_desc)
    with open(f'crawled pages/descriptions2/{file_desc}', 'r') as f:
        pages_list_desc = json.loads(f.read())['pages_description']
    for page_desc in pages_list_desc:
        text_desc = page_desc['text'].lower()
        #print('Analyzing new text')
        # Search in pages files
        for file in files:
            if 'raw_text' in page_desc:
                break
            #print(f'   {file}')
            with open(f'crawled pages/json/{file}', 'r') as f:
                if file == '.DS_Store':
                    continue
                else:    
                    pages_list = json.loads(f.read())['pages']
                    for page in pages_list:
                        text = page['text'].lower()
                        if text_desc in text:
                            page_desc['raw_text'] = text
                            print('   TEXT FOUND')
                            #print('')
                            break


    description_json = json.dumps({'pages_description':pages_list_desc}, ensure_ascii=False).encode(encoding='utf-8')
    with open(f'crawled pages/descriptions/{file_desc}', 'a') as f:
        f.write(description_json.decode())
                    
                
                
        
        
```
